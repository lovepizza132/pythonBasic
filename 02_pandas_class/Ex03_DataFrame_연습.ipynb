```python
import pandas as pd


data=pd.read_excel('data/2016-01.xls')
data
```

#### 판다스 데이타프레임의 기본적인 정보 얻어오기

- 앞에 10개 행 보기
- 뒤에 10개 행 보기

- 판매건수에서 큰 값 순으로 10개 (nlargest()이용)
- 판매건수에서 작은 값 순으로 10개 (nsmallest()이용)


```python
data.head(10)
data.tail(10)
data['판매건수'].nlargest(10)
data.sort_values('판매건수', ascending=False).head(10)
data['판매건수'].nsmallest(10)
data.sort_values('판매건수', ascending=False).tail(10)
```

#### 행에서 데이타 추출

- 야채, 딸기, 봉투행을 추출하여 매출순으로 정렬하기 (sort_values()이용)
- 바나나행부터 딸기행까지 추출
- 야채부터 딸기행까지 추출

```python
#data[(data['상품명']=='야채')| (data['상품명']=='딸기')| (data['상품명']=='봉투')].sort_values('매출')




#data[(data['상품명']=='야채')|(data['상품명']=='딸기')|(data['상품명']=='봉투')].sort_values('매출', ascending=0)
data[data['상품명'].str.contains('야채|딸기|봉투')].sort_values('매출', ascending=0)
```
