```python
# 간단한 한글처리
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

plt.rcParams['axes.unicode_minus']=False  # 추가설정 : 폰트를 변경하면 -표시가 ㅁ으로 변경되기에 '-'를 변경하지 않도록 지정
rc('font', family='Malgun Gothic')
```

```python
# 기본적인 그래프
import matplotlib.pyplot as plt

data = [10, 15, 18, 30, 20]

#plt.plot(data); #리눅스에서는 이 코딩만으로는 안그려 질 수 있음

plt.figure(figsize=(12,6)) #그래프를 그릴 영역 만들기
plt.plot(data) # 그래프를 그리기
plt.xlabel('시간')
plt.ylabel('온도')
plt.title('샘플')
plt.show() # 그래프 보여주기
```

```python
# 1- 그래프 틀(영역)을 만들기
# 2- 그래프 그리기
# 3- 그래프 보여주기
plt.plot(data, color='hotpink', linestyle='dashed', marker='o')
```
