```python
%matplotlib inline
    
import matplotlib.pyplot as plt
import numpy as np

""" 0. 플롯 스타일 """
#plt.style.use('classic')
plt.style.use('seaborn-whitegrid') #격자무늬~~~

x = np.linspace(0,10,100)

plt.plot(np.tan(x));
```

```python
""" 1. 색상지정 """

plt.plot(np.sin(x-0), color='red'); # 색상이름지정
plt.plot(np.sin(x-1), color='b'); # 색상코드명
plt.plot(np.sin(x-2), color='#FFDD44'); # 16진수 RGB
plt.plot(np.sin(x-3), color='chartreuse'); # html 색상이름
plt.plot(np.sin(x-4), color='0.75'); # 0~1 사이 값의 회색
plt.plot(np.sin(x-5), color=(1.0, 0.5, 0.2)); # RGB값을 튜플 형식으로 지정

```

```python
""" 2. 선스타일 """

# plt.plot(np.sin(x-0), linestyle='solid');
# plt.plot(np.sin(x-1), linestyle='dashed');
# plt.plot(np.sin(x-2), linestyle='dotted');
# plt.plot(np.sin(x-3), linestyle='dashdot');


plt.plot(np.sin(x-0), linestyle='-'); # 
plt.plot(np.sin(x-1), linestyle='--'); # 
plt.plot(np.sin(x-2), linestyle=':'); # 
plt.plot(np.sin(x-3), linestyle='-.'); # 
```

## r(red) g(green) b(blue) c(cyan) m(magenta) y(yellow) k(black) 

```python
""" 3. 선과 스타일을 간결하게 (대략 읽는 수준으로 ) """

plt.plot(np.sin(x-0), 'g--');
plt.plot(np.sin(x-1), ':m');
plt.plot(np.sin(x-2), '-.k');
plt.plot(np.sin(x-3), 'r:');
```

```python
''' 4. 범례 '''

plt.plot(np.sin(x-0), 'g--', label='first');
plt.plot(np.sin(x-1), ':m', label='second');
plt.plot(np.sin(x-2), '-.k', label='third');
plt.plot(np.sin(x-3), 'r:', label='fourth');

# 범례 위치: 오른쪽 상단:1, 왼쪽상단:2, 왼쪽하단:3, 오른쪽하단:4
plt.legend(loc=4);
```
