```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

""" 1.  figure(=?)   subplot    """

fig = plt.figure(figsize=(10,5))
ax1 = fig.add_subplot(1,2,1) # 1 2 
ax2 = fig.add_subplot(1,2,2) # 1 2 
ax1.plot(np.random.randn(100));
ax2.plot(np.random.randn(100).cumsum());
```

```python
""" 2. figure   (pyplot) subplot()    """

fig = plt.figure(figsize=(10,5))
ax1 = plt.subplot(1,2,1) # 1 2 
ax2 = plt.subplot(1,2,2) # 1 2 
ax1.plot(np.random.randn(100));
ax2.plot(np.random.randn(100).cumsum());
```

```python
""" 3.  ax   """

fig, ax = plt.subplots(1,2) # figure, axes
ax[0].plot(np.random.randn(100))
ax[1].plot(np.random.randn(100).cumsum())
```

```python
''' 4.     '''

fig, ax = plt.subplots(1,2, figsize=(10,7)) # figure, axes
ax[0].plot(np.random.randn(100))
ax[1].plot(np.random.randn(100).cumsum())
```
