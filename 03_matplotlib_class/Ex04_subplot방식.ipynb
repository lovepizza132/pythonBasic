```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

""" 1.  figure(그림틀=공간?)을 만들고 거기에 subplot을 추가 하는 방식 """

fig = plt.figure(figsize=(10,5))
ax1 = fig.add_subplot(1,2,1) # 1행 2열의 첫번째
ax2 = fig.add_subplot(1,2,2) # 1행 2열의 두번째
ax1.plot(np.random.randn(100));
ax2.plot(np.random.randn(100).cumsum());
```

```python
""" 2. figure로 공간을 만들고 파이플롯(pyplot)의 subplot()으로 각각의 패널 지정 """

fig = plt.figure(figsize=(10,5))
ax1 = plt.subplot(1,2,1) # 1행 2열의 첫번째
ax2 = plt.subplot(1,2,2) # 1행 2열의 두번째
ax1.plot(np.random.randn(100));
ax2.plot(np.random.randn(100).cumsum());
```

```python
""" 3. 간단하게 ax 객체를 이용하기 """

fig, ax = plt.subplots(1,2) # figure, axes
ax[0].plot(np.random.randn(100))
ax[1].plot(np.random.randn(100).cumsum())
```

```python
''' 4. 그래프 전체 크기를 지정 '''

fig, ax = plt.subplots(1,2, figsize=(10,7)) # figure, axes
ax[0].plot(np.random.randn(100))
ax[1].plot(np.random.randn(100).cumsum())
```

## 서브플롯 개념

```python
plt.figure(figsize=(12,6))

plt.subplot(221) # 2행 2열의 첫번째 영역
plt.subplot(222) # 2행 2열의 두번째 영역
plt.subplot(212) # 2행 1열의 두번째 영역

plt.show()
```

```python
plt.figure(figsize=(12,6))

plt.subplot(2,1,1) # 2행 2열의 첫번째 영역
plt.subplot(2,3,(4,5)) # 2행 2열의 두번째 영역
plt.subplot(2,3,6) # 2행 1열의 두번째 영역

plt.show()
```
