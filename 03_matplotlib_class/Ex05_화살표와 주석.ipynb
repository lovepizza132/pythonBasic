#### 차트상에서 주석 처리하는 함수 : annotate(s, xy, xytext,  arrowprops, . . . )
    + s: 주석
    + xy : 화살표시작위치
    + xytext : 주석 텍스트의 시작 위치
    + arrowprops : 화살표의 속성들
    

[참고] Matplotlib에서 화살표를 그리는 것은 plt.arrow()가 있지만 사용하기 어렵다.
    게다가 플롯의 사로 세로 비율이 바뀔 때마다 원하는 결과 얻기 어렵다.
    plt.annotate() 함수 이용을 권장한다


```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt 

fig, ax = plt.subplots()
x = np.linspace(0,20,100) #0~20 사이에 균일한 100개의 데이터

ax.plot(x, np.cos(x));
ax.axis('equal');

ax.annotate('My comment', xy=(6,2), xytext=(10,4), arrowprops=dict(facecolor='red', shrink=0.05))

ax.annotate('Important thing', xy=(6,0), xytext=(4,-6), arrowprops=dict(facecolor='red', shrink=0.05))

ax.annotate('My comment', xy=(20,1), xytext=(10,-6), arrowprops=dict(arrowstyle='->', connectionstyle='angle3, angleA=0, angleB=-90'))


```
