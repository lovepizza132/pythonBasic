## [예] 경기도에서 교통사망사고가 많은 5지역을 분석하여 도식화

(1) 발생지시도가 경기도 데이타셋 필터링

(2) 경기도내에서도 발생지시군구의 그룹핑

(3) 사망자수의 합계로 차트 만들기


```python

%matplotlib notebook
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

# 한글깨짐처리
font_loc = "c:/Windows/Fonts/malgun.ttf"
font_name=font_manager.FontProperties(fname=font_loc).get_name()
rc('font', family=font_name)
```

```python
# 데이타 읽어오기
data=pd.read_csv('data/accidentdata.csv')
data
# (1) 발생지시도가 경기도인 필터링
data2=data.loc[data['발생지시도']=='경기']
data2
```

```python
# (2) 경기도내에서도 발생지시군구의 그룹핑
data3 = data2.groupby('발생지시군구').sum()

# data3 = data2.groupby('발생지시군구').sum().sort_values('사망자수', ascending=False).head(5)
# data4=data3['사망자수']
# plt.figure(figsize=(12, 5))
# data4.plot(kind='bar')
```

```python
# (3) 사망자수의 합계로 차트 만들기
data4=data3['사망자수']
plt.figure(figsize=(12, 5))
data4.plot(kind='bar')
```

```python
# (4) 경기도 내에서 사망자 순 5개 지역 추출
data3 = data.loc[data['발생지시도']=='경기'].groupby('발생지시군구').sum().sort_values('사망자수', ascending=False).head(5)
data4=data3['사망자수']
plt.figure(figsize=(10, 5))
data4.plot(kind='bar')
```
