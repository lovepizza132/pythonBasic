##  파산한 은행 데이타집합에서 분기별 은행수를 계산하기


    1. 먼저 데이타 구조를 파악하기

    2. 'Closing Date' 날짜 항목을 datetime 오브젝트로 변기하기

    3. 'Closing Date' 날짜 정보에서 dt 접근자를 이용하여 연도(year)와 분기(quarter)를 구하기

    4.  연도별 파산은행 수 구하기

    5. 연도별, 분기별 파산은행 수 구하기

    6. 정보 시각화
    
    
    [도서] Do it 데이타분석을 위한 판다스 입문

```python
import pandas as pd

banks = pd.read_csv('./data/banklist.csv', parse_dates=['Closing Date', 'Updated Date'])


# (1) 먼저 데이타 구조를 파악하기
#banks.head()
#banks.info()

# (2) Closing Date를 datetime 오브젝트로 변경
#1번째 줄에서 함

# (3) 날짜 정보에서 dt 접근자를 이용하여 연도(year)와 분기(quarter)를 구한다
banks['Closing Year'] = banks['Closing Date'].dt.year
banks['Closing Quarter'] = banks['Closing Date'].dt.quarter
#banks.head()


# (4) 연도별 파산은행 수 구하기 count(), sum(), size() 사용하기
#print('count', banks.groupby('Closing Year').count()) # 항복별 전체 개수
#print('sum', banks.groupby('Closing Year').sum()) # 합산
#print('size', banks.groupby('Closing Year').size()) # 해당 컬럼 count
closing_year = banks.groupby('Closing Year').size()
closing_year
# (5) 연도별, 분기별 파산은행 수 구하기
closing_quarter = banks.groupby([banks['Closing Year'], banks['Closing Quarter']]).size()

#closing_year
#banks
#closing_quarter
#type(closing_quarter)
```

```python
""" 결과를 시각화 ( 간단하게 )"""
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(30,10))

ax1 = fig.add_subplot(1,2,1)
ax1 =closing_year.plot(kind='bar');

ax2 = fig.add_subplot(1,2,2)
ax2 =closing_quarter.plot(kind='area')
 
#graph1=plt.bar(x=closing_year.loc[], height=closing_year['Closing Year'])

#ax1.plot(closing_year);
#ax2.plot(closing_quarter)

```
