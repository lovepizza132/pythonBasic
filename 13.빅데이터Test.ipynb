```python
from scipy import stats
import matplotlib.pyplot as plt
import numpy as np

# 온도 X 
X = [24,25,26,27,28,29,30,31,35]
# 아이스크림 판매량 Y
Y = [22,25,28,30,33,34,36,32,38]

slope, intercept, r_value, p_value, stderr = stats.linregress(X,Y)
plt.scatter(X,Y)
plt.xlabel('기온', fontsize=22)
plt.ylabel('아이스크림 판매량', fontsize=22,)
plt.grid()

x1 = np.array(X)
plt.plot(x1, x1*slope+intercept, color='red')


print('분석결과 : ==============================')
print('r_value(상관계수): ',r_value)
if p_value<0.05:
    print('유의미한 결과')
else:
    print('유의미하지 않은 결과')
```
